#Naïve Bayes

#Imports


```python
#Load of ionosphere data set
import matplotlib.pyplot as  plt
import pandas as pd
import numpy as np
from sklearn.utils.validation import column_or_1d
from sklearn.feature_selection import VarianceThreshold
from sklearn.naive_bayes import GaussianNB       ### Because continuous data
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score, train_test_split
import sklearn.neighbors as nb
from sklearn.model_selection import cross_val_predict  
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import sklearn.model_selection as cv 
```

#Prep 


```python
from IPython.display import display, HTML

def reportClassesDataframe(report):
    report = [x.split(' ') for x in report.split('\n')]
    header = ['Class Name']+[x for x in report[0] if x!='']
    values = []
    for row in report[1:-5]:
        row = [value for value in row if value!='']
        if row!=[]:
            values.append(row)
    df = pd.DataFrame(data = values, columns = header)
    return df

def reportAccuracyDF(report):
  report = [x.split('      ') for x in report.split('\n')]
  values = []
  report[6].pop(2)
  df = pd.DataFrame(report[6:9])
  return df

def prettyReport(report):
    df = reportClassesDataframe(report)
    df2 = reportAccuracyDF(report)
    display(HTML(df.to_html(index=False,justify="center")))
    print("\n")
    display(HTML(df2.to_html( header=False,index=False,justify="center")))
    df
    df2
    return
```

```python
#Read num_rows_to_read first rows
num_rows_to_read = 2000000
```

##Balanced Dataset without Mean Encoding



```python
X_b = pd.read_csv("train_values_Prep_Bal.csv",nrows=num_rows_to_read)
y_b = pd.read_csv("train_labels_Prep_Bal.csv",nrows=num_rows_to_read)

(X_train_b, X_test_b,  y_train_b, y_test_b) = train_test_split(X_b, y_b, test_size=.3, random_state=5)
```

###Prediction

```python
clf = GaussianNB()

clf.fit(X_train_b,y_train_b)

y_pred_b = clf.predict(X_test_b)
 

print(confusion_matrix(y_test_b, y_pred_b))
print(accuracy_score(y_test_b, y_pred_b))
prettyReport(classification_report(y_test_b, y_pred_b))
```

In the case of balancing the data using undersampling, we can observe that the minority class, in our case class 1, is where a considerable improvment can be found about the other predictions. The improvment is still not the optimum but it is close to being a good prediction for this minority class

###Cross Validation


```python
cv = StratifiedKFold(n_splits=10, random_state=30, shuffle =True) 

clf = GaussianNB()
y_np = y_b.to_numpy()
y_np = y_np.ravel()
cv_scores = cross_val_score(clf,X=X_b,y=y_np,cv=cv)
np.mean(cv_scores)

predicted = cross_val_predict(clf, X=X_b, y=y_np,  cv=cv)  

print(confusion_matrix(y_np, predicted))
print(accuracy_score(y_np, predicted))
prettyReport(classification_report(y_np, predicted))
```

Performing cross validation in the case of the balanced data, we can observe a minimal variance between the prediction of the classes. The accuracy is the same. 
The difference is so insignificant that we reach the same conclusion as without cross validation.
